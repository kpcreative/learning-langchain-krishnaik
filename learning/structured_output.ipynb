```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
chat_model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
from pydantic import BaseModel, Field
class Movie(BaseModel):
    title: str = Field(..., description="The title of the movie")
    release_year: int = Field(..., description="The year the movie was released")
    genre: str = Field(..., description="The genre of the movie")
    
```

```python
# model response without structured output
response = chat_model.invoke("What is the movie Inception about?")
print(response)
```

```python
# noww with structured output
chat_model_with_structured_output = chat_model.with_structured_output(Movie)
response_2 = chat_model_with_structured_output.invoke("What is the movie Inception about?")
print(response_2)

```

```python
# if hame strucutred output and raw message ek sath dekhna hai to we can do that also
from pydantic import BaseModel, Field
class Movie(BaseModel):
    title: str = Field(..., description="The title of the movie")
    release_year: int = Field(..., description="The year the movie was released")
    genre: str = Field(..., description="The genre of the movie")
chat_model_with_structured_output = chat_model.with_structured_output(Movie, include_raw=True)
response_3 = chat_model_with_structured_output.invoke("What is the movie Inception about?")
response_3

```

```python
# nested structure ke sath try krte hai

class Actor(BaseModel):
    name:str = Field(..., description="The name of the actor")
    age:int = Field(..., description="The age of the actor")

class MovieWithActors(BaseModel):
    title: str = Field(..., description="The title of the movie")
    release_year: int = Field(..., description="The year the movie was released")
    genre: str = Field(..., description="The genre of the movie")
    actors: list[Actor] = Field(..., description="List of actors in the movie")   
    # above is the actors list of actor jo ki previous class Actor se bana hai..and that also we took list of it

chat_model_nested_struct = chat_model.with_structured_output(MovieWithActors)
response_4 = chat_model_nested_struct.invoke("What is the movie Inception about?")
response_4
# you can see response_4 me nested structure me data aa rha hai..actors ke andar actors ki list aa rhi hai..and uske andar name and age aa rha hai..
    
```
