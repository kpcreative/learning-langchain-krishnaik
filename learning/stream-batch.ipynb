```python
import os
from dotenv import load_dotenv
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python

from langchain_groq import ChatGroq

model=ChatGroq(model_name="llama-3.3-70b-versatile")
model

```

```python
# lets invoke the above llm model

model.invoke("what is capital of india?")
```

```python
response=model.stream("what is capital of india?")
response
```

```python
# you can see above that stream ne output nhi blki object i.e iterator return kiya
# now for seeing the text wee need to iterate using for loop on the response given by STREAM()

for chunk in response:
    print(chunk)
```

```python
# above you can see the response is coming in chunks as it is streaming response.
for chunk in response:
    print(chunk.content,flush=True)
    
```

```python
answer=model.batch([
    "what is capital of india?",
    "what is capital of USA?",
    "what is capital of UK?"
],

# we can also pass config for batch request as well, here max_concureency=2 mean ki at a time 2-2 request will go to llm and process hoga, isse response time kam hoga as 3 request ek sath nhi jayegi llm ko.
    config={
        "max_concurrency": 2,
    }
)

answer
```
