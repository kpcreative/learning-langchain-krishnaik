```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage
# 1. HumanMessage: represents a message from a human user. It typically contains the text input provided by the user.
# 2. AIMessage: represents a message generated by an AI model in response to a human message.
# 3. SystemMessage: represents a message that provides instructions or context to the AI model. It can be used to set the behavior or constraints for the AI's response.

messages=[
    SystemMessage(content="You are a helpful poetry expert assistant."),
    HumanMessage(content="Write a poem about the sea.")
]

# we will pass this list of messages to the chat model, and it will generate a response based on the conversation history and the instructions provided in the system message.
response = chat_model.invoke(messages)
print(response)
```
